```python
import pandas as pd
portfolio = pd.read_csv("Portfolio Data.csv", encoding = "latin-1")
```

```python
# Calculate Open Seats
portfolio['Open Seats'] = portfolio['Desired Enrollment'] - portfolio['Students']

# Make Zip + Grade key 
portfolio['zipgrade'] = portfolio['Facility: Zip/Postal Code'].astype(str) + portfolio['Grade Level: Grade']
```

```python
portfolio.groupby('Accountability Record: Site Name').sum('Students')
```

```python
portfolio.groupby('Facility: Zip/Postal Code').sum('Students')
```

```python
portfolio.groupby('zipgrade').sum('Students').sort_values('Facility: Zip/Postal Code')
```
